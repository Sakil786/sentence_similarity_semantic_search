```python
import pandas as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import InputExample, InputFeatures
from sentence_transformers import SentenceTransformer, InputExample, losses
import pandas as pd
from sentence_transformers import SentenceTransformer, InputExample
from torch.utils.data import DataLoader

model = SentenceTransformer('distilbert-base-nli-mean-tokens')
```

```python
# Load news data
train_data = pd.read_csv("/content/ready_data.csv")
```

```python
train_data.head()
```

```python
train_data.columns=['text_a','text_b','label']
```

```python
train_data['label']=train_data['label'].astype('float')
```

```python
train_data.dropna(inplace=True)
```

```python
train_data.head(2)
```

```python
# Create InputExamples
train_examples = []
for i, row in train_data.iterrows():
    text_a = row['text_a']
    text_b = row['text_b']
    label = row['label']
    example = InputExample(texts=[text_a, text_b], label=label)
    train_examples.append(example)
```

```python
# Define the training loss
train_loss = losses.CosineSimilarityLoss(model=model)

# Define the training dataloader
train_dataloader = DataLoader(train_examples, batch_size=32, shuffle=True)
```

```python
# Fine-tune the model
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=5, warmup_steps=100)
```

```python
model.save("/content/sentence_similarity_semantic_search")
```
